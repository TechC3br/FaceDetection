```python
import cv2
import mediapipe as mp
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model(r'Downloads/face_direction_mobilenet.h5')

# Initialize Mediapipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=False, max_num_faces=1, min_detection_confidence=0.5)

# Initialize OpenCV for video capture
cap = cv2.VideoCapture(0)  # Use 0 for the default camera

# Define a dictionary for face orientation classes
orientation_dict = {0: 'Left', 1: 'Right', 2: 'Up', 3: 'Down'}

# Define the input size expected by the model
input_size = (224, 224)

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        break
    
    # Convert the frame to RGB for Mediapipe
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Get face mesh landmarks
    result = face_mesh.process(rgb_frame)
    
    if result.multi_face_landmarks:
        for face_landmarks in result.multi_face_landmarks:
            # Get face bounding box or region of interest (ROI)
            h, w, _ = frame.shape
            x_min, x_max, y_min, y_max = w, 0, h, 0
            for lm in face_landmarks.landmark:
                x, y = int(lm.x * w), int(lm.y * h)
                x_min, x_max = min(x_min, x), max(x_max, x)
                y_min, y_max = min(y_min, y), max(y_max, y)
            
            # Extract face region
            face_roi = frame[y_min:y_max, x_min:x_max]

            if face_roi.size > 0:
                # Resize the face ROI to the input size of the model
                face_resized = cv2.resize(face_roi, input_size)
                face_resized = np.expand_dims(face_resized, axis=0)  # Add batch dimension

                # Normalize the input using preprocess_input (as in training)
                face_resized = face_resized / 255.0  # Simple normalization

                # Predict face orientation
                prediction = model.predict(face_resized)
                predicted_class = np.argmax(prediction)  # Get the class with highest probability
                
                # Get the orientation label
                orientation_label = orientation_dict[predicted_class]

                # Display the predicted orientation on the frame
                cv2.putText(frame, f"Orientation: {orientation_label}", (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Show the frame with the prediction
    cv2.imshow('Face Orientation Detection', frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
